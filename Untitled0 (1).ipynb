```python
def preprocess_whiteboard(path):

    img = cv2.imread(path)

    # 1️⃣ upscale (super important)
    img = cv2.resize(img, None, fx=3, fy=3, interpolation=cv2.INTER_CUBIC)

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # 2️⃣ contrast boost
    clahe = cv2.createCLAHE(clipLimit=4.0, tileGridSize=(8,8))
    gray = clahe.apply(gray)

    # 3️⃣ smooth but KEEP EDGES (much better than NLM)
    gray = cv2.bilateralFilter(gray, 9, 75, 75)

    # 4️⃣ Otsu threshold (stronger than adaptive for whiteboard)
    _, th = cv2.threshold(
        gray, 0, 255,
        cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU
    )

    # 5️⃣ thicken strokes (CRUCIAL for pix2tex)
    kernel = np.ones((2,2), np.uint8)
    th = cv2.dilate(th, kernel, iterations=1)

    return th
```

```python
ppimg=preprocess_whiteboard(image_path)
```

```python
cv2.imwrite("preprocessed.png", ppimg)
```
